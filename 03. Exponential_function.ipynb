# ðŸ“— Exponential Functions - Basic Concepts and Visualisation 

## 1. Introduction

Exponential functions have the form f(x) = a^x, where the base is a a positive number not equal to 1.

- a > 1 -> Increasing
- 0 < a < 1 -> Decreasing

## ðŸ§ª Step 2. Exponential Functions - Basic Concepts and Visualisation

In this step, we will visualize the basic exponential function

**f(x) = 2^x**, where the base a = 2.

This function demonstrates exponential growth.
As x increases, f(x) increases rapidly. As x decreases, the values approach 0. 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
y = 2 ** x 

plt.figure(figsize = (8, 5))
plt.plot(x, y, label = "f(x) = 2^x", color = 'green')
plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Graph of the Basic Exponential Function: f(x) = 2^x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“‰ Step 3. Visualising Changes in Base a 

In this step, we will compare the graphs of the exponential function

**f(x) = a^x** using various values of base a.

We will visualize:

- a > 1 (e.g., a = 2, 3) : exponential growth
- 0 < a < 1 (e.g., a = 0.5): exponential decay
- a = 1 : constant function f(x) = 1

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
a_values = [0.5, 1, 2, 3]

plt.figure(figsize = (8, 5))

for a in a_values:
    y = a**x
    plt.plot(x, y, label=f"a = {a}")

plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Effect of base a on f(x) = a^x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“‰ Step 4. Comparing f(x) = a^x and f(x) = a^(-x)

In this step, we compare the exponential function f(x) = a^x with its mirroed version f(x) = a^(-x).

This helps us understand how negating the exponent reflects the curve across the y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
a = 2

y1 = a**x
y2 = a**(-x)

plt.figure(figsize=(8,5))
plt.plot(x, y1, label = "f(x) = 2^x", color = 'blue')
plt.plot(x, y2, label = "f(x) = 2^(-x", color = 'red')

plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Comparison of f(x) = a^x and f(x) = a^(-x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“ˆ Step 5. Plotting the Natural Exponential Function f(x) = e^x

In this step, we will visualize the exponential function with base \( e \ ), known as the natural exponential function:

**f(x) = e^x**, where \( e â‰ˆ 2.718 \ )

This function grows continuously and is used in many fields such as finance, biology, and calculus. 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
y = np.exp(x) # f(x) = e^x

plt.figure(figsize = (8,5))
plt.plot(x, y, label = 'f(x) = e^x', color = 'green')

plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Graph of the Nautral Exponential Function : f(x) = e^x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
