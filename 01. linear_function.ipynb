# ðŸ“— Linear Functions - Basic Concepts and Visualisation

## 1. Introduction 

Linear functions are the most basic type of function, represented by a straight line on a graph. 
It is defined by the formula:

f(x) = a*x + b

- 'a' is the slope : how steep the line is.
- 'b' is the y-intercept: the value of y when x = 0.

## ðŸ§ª Step 2. Code Cell - Preparing for Visualisation with Python

### ðŸ“Œ 2.1 Importing Modules and Creating x-values

We will use 'numpy' to create data and 'matplotlib' to draw graphs.
Let's start by creating 100 values between -10 to use as the x-axis. 

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-10, 10, 100)
```

### ðŸ“Œ 2.2 Plotting a Basic Linear Function 

Now we will choose a slope ('a') and intercept ('b') to plot the most basic linear function. 
In this example, we visualise the function \( f(x) = 2x+1).

```python
a = 2 # slope
b = 1 # intercept
y = a*x + b
```

```python
plt.figure(figsize = (8,5))
plt.plot(x, y, label = f"f(x) = {a}x + {b}", color = 'blue')
plt.axhline(0, color = 'black', linewidth = 0.5) # xì¶•
plt.axvline(0, color = 'black', linewidth = 0.5) # yì¶•
plt.grid(True)
plt.legend()
plt.title("Graph of f(x) = ax + b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“ˆ Step 3. Visualising Changes in Slope (a)

```python
plt.figure(figsize = (8,5))

for a in [-2, -1, 0, 1, 2]:
    y = a*x 
    plt.plot(x, y, label = f"a = {a}")

plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Effect of Slope a on f(x) = ax")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
