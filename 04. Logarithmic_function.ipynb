# üìó Logarithmic Functions - Basic Concepts and Visualisation 

## 1. Introduction

Logarithmic functions are the inverse of exponential functions.

They play a crucial role in mathematics, especially in solving euqations involving exponential growth or decay.

A logarithmic function answers the question:

"To what power must we raise a given base to get a certain number?"

For example:

\[\log_2{8} = 3 \quad \text{because} \quad 2^3 = 9\]

## üóíÔ∏è Step 1. Understanding the Logarithmic Function Form

The general form of a logarithmic function is:

\[ f(x) = \log_a{x}\] 

where \(a > 0 \) and \(a \ne 1 \).

This function increases slowly and only exists for \(x >0 \).

## üìà Step 2. Plotting the Basic Logarithmic Function

In this step, we will visualize the basic logarithmic function 

**f(x) = log‚ÇÇx**, where the base is 2.

The graph only exists for X > 0 and increases slowly.

It passes through the point (1, 0), because log‚ÇÇ(1) = 0.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 10, 400)  # x should be bigger than 0!
y = np.log2(x)  # log base 2

plt.figure(figsize=(8,5))
plt.plot(x, y, label="f(x) = log‚ÇÇx", color='blue')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title("Graph of the Basic Logarithmic Function: f(x) = log‚ÇÇx")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## üìà Step 3. Visualising Changes in Base a

In this step, we compare logarithmic functions with different base values:

**f(x) = log‚Çêx**, where a = 2, 5, 10.

As the base increases:
- The graph becomes flatter (increases more slowly).
- All curves pass through the point (1, 0).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 10, 400)
base_values = [2, 5, 10]

plt.figure(figsize = (8,5))

for a in base_values:
    y = np.log(x) / np.log(a) # log_a(x) = log(x) / log(a)
    plt.plot(x, y, label = f"log base {a}")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title("Effect of Base a on f(x) = log‚Çêx")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## üìâ Step 4. Comparing Logarithmic and Exponential Functions

In this step, we will compare the logarithmic function

**f(x) = log‚ÇÇx** and its inverse function, the exponential function **f(x) = 2^x**.

These two functions are reflections of each other across the line **y = x**.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.01, 10, 400)  
x2 = np.linspace(-3, 3, 400)

log_y = np.log(x1)
exp_y = 2**x2

plt.figure(figsize=(8, 5))
plt.plot(x1, log_y, label="f(x) = log‚ÇÇx", color='blue')
plt.plot(x2, exp_y, label="f(x) = 2^x", color='green')
plt.plot(x1, x1, label="y = x", color='gray', linestyle='--')  # ÎåÄÏπ≠ Í∏∞Ï§ÄÏÑ†

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title("Logarithmic and Exponential Functions as Inverses")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
