# ðŸ“— Quadratic Functions - Basic Concepts and Visualisaton

## 1. Introduction

Quadratic functions are a fundamental type of nonlinear function that appear as curves (parabolas) on a graph.
They are defined by the formula:

*f(x) = ax^2 + bx + c*

- 'a' determines the direction and width of the curve.
- 'b' moves the vertex left or right.
- 'c' determines the y-intercept (the value of y when x = 0).

## ðŸ§ª Step 2.Plotting a Basic Quadratic Function


Now, we will plot the most basic form of a quadratic function: \( f(x) = x^2)\.
This version has no slope or intercept, so we can observe the pure shape of the curve.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y = x**2

plt.figure(figsize = (8,5))
plt.plot(x, y, label = "f(x) = x**2", color = 'purple')
plt.axhline(0, color ='black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Graph of f(x) = x**2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“ˆ Step 3. Visualising Changes in Coefficient a 

The coefficient 'a' in a quadratic function determines **the direction and steepness of the parabola**:

- 'a > 0' : The parabola opens upwards (smiile shape)
- 'a < 0' : The parabola opens downwards (frown shape)
- Larger '|a|' : The parabola becomes narrower and steeper.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

a_values = [-2, -1, 0.5, 1, 2]

plt.figure(figsize = (8,5))
for a in a_values:
    y = a * x**2 
    plt.plot(x, y, label=f"a = {a}")


plt.axhline(0, color = 'black', linewidth=0.5)
plt.axvline(0, color = 'black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title("Effect of Coefficient a on f(x) = ax**2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“ˆ Step 4. Visualising Changes in Coefficien b

The coefficient 'b' in a quadratic function affects **the horizontal (left/right) position of the vertex**.

- Changing 'b' moves the parabola **sideways**, without changing its shape or direction.
- The x-coordinate of the vertex is calculated by the formula \(x = \frac{-b}{2a} \).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

a = 1
c = 0
b_values = [-6, -2, 0, 2, 4]

plt.figure(figsize=(8,5))

for b in b_values:
    y = a*x**2 + b*x + c
    plt.plot(x, y, label = f"b = {b}")

plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Effect of Coefficient b on f(x) = ax^2 + bx + c")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“ˆ Step 5. Visualising Changes in Constant Term c

The constant term c in a quadratic function determines the vertical (up/down) shift of the graph.
- Increasing c moves the parabola up, while decreasing it moves it down.
- The shape and vertex position remain unchanged.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

a = 2
b = 0
c_values = [-4, -1, 0, 2, 5]

plt.figure(figsize=(8,5))

for c in c_values:
    y = a*x**2 + b*x + c
    plt.plot(x, y, label = f"c = {c}")


plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Effect of Constant c on f(x) = ax^2 + bx + c")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## ðŸ“ˆ Step 6. Visualising the Combined Effect of a, b, and c

The three coefficients 'a', 'b', and 'c' in a quadratic function each affect the graph:

- 'a' : determines the direction (up/down) and width of the curve
- 'b' : shifts the vertex horizontally (left/right)
- 'c' : moves the graph up or down

In this step, we visualise various combinatons of '(a, b, c)'
to **compare their effects on the parabola**.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

params = [ (1, 0, 0), (1, 2, -3), (-1, 0, 0), (-1, -2, 3), (0.5, 4, -2)]

plt.figure(figsize=(10, 6))

for a, b, c in params:
    y = a*x**2 + b*x + c
    plt.plot(x, y, label = f"a = {a}, b = {b}, c = {c}")

plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.title("Combined Effects of a, b, and c on f(x) = ax^2 +bx + c")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## ðŸ“ Step 7. Visualising the Vertex of a Quadratic Function

The vertex of a quadratic function \(f(x) = ax^2 + bx + c \)
can be found using this formula:

\[x = \frac{-b}{2a}, \quad y = f(x)\]

In this step, we calculate the vertex for each function and **mark it as a red dot on the graph**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

params = [(1,2,-3), (-1, -2, 1), (0.5, 4, -2)]

plt.figure(figsize = (10, 6))

for a, b, c in params:
    y = a*x**2 + b*x + c
    vertex_x = -b/(2*a)
    vertex_y = a*vertex_x**2 + b*vertex_x + c

    plt.plot(x, y, label = f"a ={a}, b={b}, c={c}")
    plt.plot(vertex_x, vertex_y, 'ro')


plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend
plt.title("Vertex of f(x) = ax^2 + bx + c")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## ðŸ§  Step 6. Visualising the Discriminant D

In a quadratic function, the **discriminat** is calculated as:

\[D = b^2 - 4ac\]

- 'D > 0' : Two distinct real roots -> Graphs intersects the x-axis at **two points**
- 'D = 0' : One real root (repeated) -> Graph **touches** the x-axis at **one pont** (the vertex)
- 'D < 0' : No real root -> Graph **does not intersect** the x-axis

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

params = [(1, 4, 1), (1, 2, 1), (1, 1, 1)]

plt.figure(figsize=(10, 6))

for a, b, c, in params:
    y = a*x**2 + b*x + c
    D = b**2 - 4*a*c
    label = f"D={D} (a={a}, b={b}, c={c})"
    plt.plot(x, y, label = label)

plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.title("Effect of Dsicrimant D on f(x) = ax^2 + bx + c")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
