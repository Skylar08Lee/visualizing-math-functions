# üìó Comparing Different Functions

## 1. Introduction 

In this chapter, we compare the overall shapes, growth patterns, and key features of different types of functions:

- Linear (f(x) = ax + b)
- Quadratic (f(x) = ax¬≤ + bx + c)
- Exponential (f(x) = a^x)
- Logarithmic (f(x) = log‚Çêx)

By plotting them together or side by side, we can better understand how they behave differently.

## üìâ Step 2. Comparing Linear and Quadratic Functions

In this step, we compare a linear function (f(x) = x) and a quadratic function (f(x) = x¬≤ ) on the same graph.

This help us understnad the difference between straight-line growth and curved growth.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

linear_y = x             # f(x) = x
quadratic_y = x ** 2     # f(x) = x^2

plt.figure(figsize=(8, 5))
plt.plot(x, linear_y, label="f(x) = x", color='blue')
plt.plot(x, quadratic_y, label="f(x) = x¬≤", color='red')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.title("Linear vs Quadratic Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## üìâ Step 3. Comparing Exponential and Logarithmic Functions

In this step, we compare the exponential function \(f(x) = 2^x \) and the logarithmic function \(f(x) = \lo_2{x} \).

These two functions are **inverse of each other** and are symmetric across the line \(y = x \).

```python
import numpy as np
import matplotlib.pyplot as plt

# x Í∞í ÏÑ§Ï†ï
x1 = np.linspace(0.1, 10, 400)   # Î°úÍ∑∏ Ìï®ÏàòÏö© (0Î≥¥Îã§ Ïª§Ïïº Ìï®)
x2 = np.linspace(-3, 3, 400)     # ÏßÄÏàò Ìï®ÏàòÏö©

# Ìï®Ïàò Ï†ïÏùò
log_y = np.log2(x1)              # f(x) = log‚ÇÇx
exp_y = 2 ** x2                  # f(x) = 2^x

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(8, 5))
plt.plot(x1, log_y, label="f(x) = log‚ÇÇx", color='blue')
plt.plot(x2, exp_y, label="f(x) = 2^x", color='green')
plt.plot(x1, x1, label="y = x", color='gray', linestyle='--')  # ÎåÄÏπ≠ÏÑ†

# Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Exponential vs Logarithmic Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## üß™ Step 4. Comparing All Function Types Together

In this step, we plot and compare the main functions we've studied:

- Linear: \(f(x) = x \)
- Quadratic: \(f(x) = x^2 \)
- Exponential: \(f(x) = 2^x \)
- Logarithmic: \(f(x) = \log_2{x} \)

This helps us visually understand the different **growth patterns** and **shapes** of each function. 

```python
import numpy as np
import matplotlib.pyplot as plt

# x Î≤îÏúÑ Ï†ïÏùò
x = np.linspace(0.1, 10, 400)
x2 = np.linspace(-3, 3, 400)

# Í∞Å Ìï®Ïàò Ï†ïÏùò
linear_y = x
quadratic_y = x**2
exponential_y = 2 ** x2
logarithmic_y = np.log2(x)

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 6))
plt.plot(x, linear_y, label="Linear: f(x) = x", color='blue')
plt.plot(x, quadratic_y, label="Quadratic: f(x) = x¬≤", color='orange')
plt.plot(x2, exponential_y, label="Exponential: f(x) = 2^x", color='green')
plt.plot(x, logarithmic_y, label="Logarithmic: f(x) = log‚ÇÇx", color='red')

# Ï∂ï Î∞è Ïä§ÌÉÄÏùº ÏÑ§Ï†ï
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title("Comparison of Function Types")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-5, 25)
plt.show()
```
